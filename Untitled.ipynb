```r
library(tidyverse)
library(repr)
library(tidymodels)
library(rvest)
```

```r
url <- "http://roycekimmons.com/system/generate_data.php?dataset=exams&n=1000"
exam_data <- read_csv(url)
colnames(exam_data) <- c("gender", "race_ethnicity", "parental_level_of_education", "lunch", "test_preparation_course", 
                         "math_score", "reading_score", "writing_score")

exam_data

exam_data1<- exam_data |>
mutate(parental_level_of_education = as_factor(parental_level_of_education)) |>
mutate("mean_score" = (math_score + reading_score + writing_score)/3) |>
select(math_score, reading_score, writing_score, mean_score)

exam_data1

exam_graph <- exam_data1 |>
ggplot(aes(x = writing_score, y = reading_score)) +
geom_point() +
labs( y = "Reading Score", x = "writing Score")



exam_graph


```

```r
exam_graph2 <- exam_data |>
ggplot(aes(x = writing_score, y = math_score)) +
geom_point() +
labs( y = "Math Score", x = "writing Score")

exam_graph2
```

```r
exam_graph3 <- exam_data1 |>
ggplot(aes(x = math_score, y = reading_score)) +
geom_point() +
labs( y = "Reading Score", x = "Math Score")

exam_graph3
```

```r
install.packages("gridExtra")              
library("gridExtra")
grid.arrange(exam_graph, exam_graph2, exam_graph3, ncol = 3)  
```
