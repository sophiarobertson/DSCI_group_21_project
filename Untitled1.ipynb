Group 21 data science project proposal: <b>The Relationship Between A Students Exam Preformances</b>.

<b>Introduction</b>: Our project deals with how different factors may affect how well a student performs on an exam in various subjects as well as the relationship between a student's scores across all the subjects tested (math, reading, and writing). We were curious to know if the students seemed to perform well/poorly in only a specific subject or all three. The question we will be trying to answer in our project is "based on a students exam score in one subject (percentage grade), can we predict what their score would be in other subjects using regression?". The data set that we will be using to help answer this question consists of marks secured by US highschool students in various subjects and is used to understand the effect of the parents education background, financial ability etc on students performances. 


```r
library(tidyverse)
library(repr)
library(tidymodels)
library(rvest)
```

<b>Reading The Data From The Web:</b>

```r
url <- "http://roycekimmons.com/system/generate_data.php?dataset=exams&n=1000"
exam_data <- read_csv(url)
colnames(exam_data) <- c("gender", "race_ethnicity", "parental_level_of_education", "lunch", "test_preparation_course", 
                         "math_score", "reading_score", "writing_score")
exam_data
```

<b>Cleaning The Data Into Only The Columns We Will Be Using:<b>

```r
exam_data1<- exam_data |>
mutate(student = 1:1000) |>
mutate("mean_score" = (math_score + reading_score + writing_score)/3) |>
select(student, reading_score, writing_score)

exam_data1
```

<b>Training Data:</b>

```r
data_split <- initial_split(exam_data1, prop = 0.75, strata = reading_score)
data_train <- training(data_split)
data_test <- testing(data_split)

data_train


summary(data_train)
```

<b>Scatter Plot Of Training Data:<b/>

```r
graph1 <- ggplot(data_train, aes(x = reading_score, y = writing_score)) +
geom_point() +
xlab("Reading Score") +
ylab("writing Score")

graph1
```

<b>Methods</b>: Using the regression K-nearest neighbor model we will be predicting what a student's writing score would be on an exam, based on their reading exam score. From our data set we will use the reading_score, and writing_score columns to do this. Using these variables we will compare them in a scatter plot to further visualize the relationship between the students' exam scores in the different subjects.

<b>Expected outcomes and significance:</b> We expect to find a strong correlation between the reading and writing scores of students in the dataset. When given a particular reading score, we may be able to predict a writing score based on the regression line of the data set. In the future this can help understand the relationship between a student's overall school performance and their individual subject performances. This is an interesting topic in that it could lead to future questions like "if a student excels in one subject, will you see the trend in their other subjects?".
