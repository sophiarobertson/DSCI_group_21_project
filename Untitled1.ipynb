Introduction: This data set consists of the marks secured by the students in various subjects and is used to understand the effect of the parents education background, financial ability etc on students performances. The question we will be trying to answer in out project "based on a students exam score in one subject (percentage grade) can we predict what their score would be in other subjects using regression". 


```r
library(tidyverse)
library(repr)
library(tidymodels)
library(rvest)
```

```r
url <- "http://roycekimmons.com/system/generate_data.php?dataset=exams&n=1000"
exam_data <- read_csv(url)
colnames(exam_data) <- c("gender", "race_ethnicity", "parental_level_of_education", "lunch", "test_preparation_course", 
                         "math_score", "reading_score", "writing_score")
exam_data
```

```r
exam_data1<- exam_data |>
mutate(student = 1:1000) |>
mutate("mean_score" = (math_score + reading_score + writing_score)/3) |>
select(student, reading_score, writing_score)

exam_data1
```

```r
data_split <- initial_split(exam_data1, prop = 0.75, strata = reading_score)
data_train <- training(data_split)
data_test <- testing(data_split)

data_train
data_test

summary(data_train)
```

```r
graph1 <- ggplot(data_train, aes(x = reading_score, y = writing_score)) +
geom_point() +
xlab("Reading Score") +
ylab("writing Score")

graph1
```

Methods: Using the regresion K-nearest neighbor model we will be predicting what a students writing score would be on an exam, based off their reading exam score. From our data set we will use the reading_score, and writing_score columns to do this. Using these variables we will compare them in a scatter plot to further vizualize the realationship between the students exam scores in the different subjects.

Expected outcomes and significance: We expect to find a strong correlation between the reading and writing scores of students in the dataset. When given a particular reading score, we may be able to predict a writing score based on the regression line of the data set. In the future this can help understand the relationship between a students overall school prefomace and their individual subject preformances. This is an interesting topic in that it could lead to future question like "if a student excels in one subject, will you see the trend in their other subjects?".
